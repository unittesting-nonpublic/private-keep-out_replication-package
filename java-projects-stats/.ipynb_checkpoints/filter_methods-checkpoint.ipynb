```python
# DO NOT RUN!!!! FILTER NOT INCLUDED PROJECTS 

import glob
import os
import subprocess
import pandas as pd


df = pd.read_csv('methods_invoked_grouped.tsv.zip', delimiter='\t',low_memory=False)
with open('final_grouped14.tsv', 'w') as tofile:
    for index, row in df.iterrows():
        search_string = row['Method Name'].split("(")[0].split(".")[len(row['Method Name'].split("(")[0].split("."))-1].split("$")[len(row['Method Name'].split("(")[0].split(".")[len(row['Method Name'].split("(")[0].split("."))-1].split("$"))-1]
        # if len(row['Method Name'].split("(")[1]) == 1:
        #     search_string = search_string + "()"
        # print(search_string)
        java_files = glob.glob("/Volumes/2TB_EXT/projects_tmp/" + row['Project Module'] + "/" + '**/' + row['Internal Test Case'].split("(")[0].split(".")[len(row['Internal Test Case'].split("(")[0].split("."))-2].split("$")[0] + ".java", recursive=True)
        if len(search_string) > 0:
            for file in java_files:
                found = False
                with open(file, 'r') as file:
                    test_found = False
                    ix = 0
                    test_string = row['Internal Test Case'].split("(")[0].split(".")[len(row['Internal Test Case'].split("(")[0].split("."))-1].split("$")[len(row['Internal Test Case'].split("(")[0].split(".")[len(row['Internal Test Case'].split("(")[0].split("."))-1].split("$"))-1] + "("
                    for line in file:
                        if test_string in line and test_found == False:# and '{' in line:
                            test_found = True
                            if '{' in line:
                                ix = ix + 1
                                continue

                        if test_found == True:
                            if '{' in line:
                                ix = ix + 1

                            if row['Method Name'].split("(")[1] == ")" and search_string in line and len(line.split(search_string)) > 1:
                                x = list(line.split(search_string)[1])
                                if len(x) > 1 and x[0]==("(") and x[1]==(")"):
                                    found = True
                                    break

                            if search_string + "(" in line and len(line.split(search_string)) > 1:
                                x = list(line.split(search_string)[1])
                                if len(x) > 1 and x[0]==("(") and x[1]==(")"):
                                    continue
                                else:
                                    found = True
                                    break

                            if '"' + search_string + '"' in line:
                                found = True
                                break

                            if '}' in line:
                                ix = ix - 1
                                if ix <= 0: 
                                    break
                if found:
                    row_values = '\t'.join(map(str, row.values))
                    tofile.write(row_values + '\n')
                    break
            # else:
            #     print(index, end=".")
```
